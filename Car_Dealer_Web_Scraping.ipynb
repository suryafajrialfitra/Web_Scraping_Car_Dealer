## Loop Function

```python
name = []
info = []
total_price = []
monthly_fee = []
cash_price = []
available = []

for i in range(1,9):
    
    # website 
    website = 'https://www.carsome.id/beli-mobil-bekas?pageNo=' + str(i)
    
    # request
    response = requests.get(website)
    
    # soup
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # results
    results = soup.find_all('article', {'class': 'mod-b-card'})

    for result in results:
    
        # name
        try:
            name.append(result.find('a', {'class': 'mod-b-card__title'}).get_text().strip())
        except:
            name.append('NULL')

        # info
        try:
            info.append(result.find('div', {'class': 'mod-b-card__car-other'}).get_text())
        except:
            info.append('NULL')

        # total_price
        try:
            total_price.append(result.find('strong').get_text().strip())
        except:
            total_price.append('NULL')

        # monthly_fee
        try:
            monthly_fee.append(result.find('div',{'class' : 'mod-tooltipMonthPay'}).get_text().strip())
        except:
            monthly_fee.append('NULL')

        # price_cash
        try:
            cash_price.append(result.find('div',{'class' : 'mod-card__price-cash'}).get_text().strip())
        except:
            cash_price.append('NULL')
            
        # avaliable
        try:
            available.append(result.find('p',{'class' : 'mod-b-card__mask-text'}).get_text())
        except:
            available.append('NULL')
        
        
```

## Dataframe

```python
car_dealer = pd.DataFrame({'Name':name, 'Info':info, 'Total Price':total_price, 'Monthly Fee':monthly_fee,
                           'Cash Price':cash_price,'Available':available})
```

```python
car_dealer
```

## Data Cleaning

### - Fix Name Column

```python
car_dealer['Name'].str.split('\n',5, expand = True)
```

```python
car_dealer[["Year","Brand","NULL","Name","Type","Version"]] = car_dealer['Name'].str.split('\n',5, expand = True)
```

```python
car_dealer
```

```python
car_dealer["Type"] = car_dealer["Type"].str.replace('\n','')

car_dealer = car_dealer.drop(columns=["NULL"])
```

```python
car_dealer
```

### - Fix Info Column

```python
car_dealer['Info'].str.split(' ',3, expand = True)
```

```python
car_dealer[["KM","km","Car Transmissions","Location"]] = car_dealer['Info'].str.split(' ',3, expand = True)

car_dealer = car_dealer.drop(columns=["Info","km"])
```

```python
car_dealer
```

### - Fix Monthly Column

```python
car_dealer["Monthly Fee"] = car_dealer["Monthly Fee"].str.replace('Rp','')
car_dealer["Monthly Fee"] = car_dealer["Monthly Fee"].str.replace('jt/bln','')
car_dealer["Monthly Fee"] = car_dealer["Monthly Fee"].str.replace(',','.')
car_dealer["Monthly Fee"] = car_dealer["Monthly Fee"]+"0.000"
car_dealer["Monthly Fee"] = car_dealer["Monthly Fee"].str.replace(' ','')

car_dealer
```

### - Fix Cash Price Column

```python
car_dealer["Cash Price"] = car_dealer["Cash Price"].str.replace('Rp','')
car_dealer["Cash Price"] = car_dealer["Cash Price"].str.replace('(Cash)','')

car_dealer
```

```python
car_dealer['Cash Price'].str.split(' ',2, expand = True)
```

```python
car_dealer[["Null","Cash Price","Null"]] = car_dealer['Cash Price'].str.split(' ',2, expand = True)
car_dealer
```

```python
car_dealer = car_dealer.drop(columns=["Null"])
```

```python
car_dealer
```

### - Fix Avaliable Column

```python
car_dealer["Available"] = car_dealer["Available"].str.replace("NULL","Y")
car_dealer["Available"] = car_dealer["Available"].str.replace("Unit Sudah Dipesan","N")

car_dealer
```

## Final Data

```python
car_dealer = car_dealer[['Year','Brand','Name','Type','Version','KM','Car Transmissions','Total Price','Monthly Fee','Cash Price','Location','Available']]
car_dealer 
```
